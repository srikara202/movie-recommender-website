```python
# Need to convert the above into a list of strings with only the genre names
import ast
def convert(g_list):
  g_list = ast.literal_eval(g_list)
  res = []
  for g in g_list:
    res.append(g['name'])
  return res

movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
 movies['crew'] = movies['crew'].apply(convert)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head(3)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
new_df.iloc[0]['tags']
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.replace(","," "))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.replace('"',''))
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:20]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("Spider-Man 3")
```
